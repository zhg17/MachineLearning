## Cleaning Data

### Missing Data
Remove all rows (observations) if any columns includes any missing data

```python
raw_data.dropna(how="any", axis = 0, inplace = True)
raw_data = raw_data.reset_index()
f_data.dropna(how="any", axis = 0, inplace = True)
```

### Removing all columns that contain data about each fighter's previous opponents.

We clean the data of irrelavent data such as the records of the fighters' past opponents because they do not impact the predictive power of our model.

```python
#remove all columns that contain data about each fighters previous opponents
opponents_col = []

for column in raw_data: 
    if 'opp' in column:
        opponents_col.append(column)

raw_data.drop(opponents_col, axis=1, inplace = True)
```

### Removing all columns that contain subtypes of decision victories
Remove columns 'win_by_Decision_Majority', 'win_by_Decision_Split','win_by_Decision_Unanimous' for both Red and Blue fighters as they track subtypes of decision victories even though there already exists a feature 'win_by_Decision' that counts all wins by decision for a fighter

```python
#no need to track the subtype of decision victories
red_list = ['R_win_by_Decision_Majority', 'R_win_by_Decision_Split','R_win_by_Decision_Unanimous']
blue_list = ['B_win_by_Decision_Majority', 'B_win_by_Decision_Split','B_win_by_Decision_Unanimous']

raw_data['R_win_by_Decision'] = raw_data[red_list].sum(axis=1)
raw_data['B_win_by_Decision'] = raw_data[blue_list].sum(axis=1)

raw_data.drop(red_list, axis=1, inplace = True)
raw_data.drop(blue_list, axis=1, inplace = True)
```

### Removing all columns that contain subtypes of significant strikes
significant strikes, both landed and attempted, are further broken down into subcategories by body part : BODY, LEG, HEAD and distance: CLINCH, GROUND, DISTANCE. This is unnecessary as the data is already represented by columns avg_SIG_STR_landed (Total significant strikes landed) and avg_SIG_STR_attempted (Total significant strikes attempted).


```python
to_remove = []

#significant strikes landed and attempted are further broken down into subcategories by
#body part : BODY, LEG, HEAD and distance: CLINCH, GROUND, DISTANCE

to_remove.extend([
    'B_avg_BODY_att', 'B_avg_BODY_landed','R_avg_BODY_att', 'R_avg_BODY_landed',
    'B_avg_LEG_att', 'B_avg_LEG_landed', 'R_avg_LEG_att', 'R_avg_LEG_landed',
    'B_avg_HEAD_att', 'B_avg_HEAD_landed', 'R_avg_HEAD_att', 'R_avg_HEAD_landed'
    ])

to_remove.extend([
    'B_avg_CLINCH_att', 'B_avg_CLINCH_landed','R_avg_CLINCH_att', 'R_avg_CLINCH_landed',
    'B_avg_GROUND_att', 'B_avg_GROUND_landed', 'R_avg_GROUND_att', 'R_avg_GROUND_landed',
    'B_avg_DISTANCE_att', 'B_avg_DISTANCE_landed', 'R_avg_DISTANCE_att', 'R_avg_DISTANCE_landed'
    ])

raw_data.drop(to_remove, axis=1, inplace = True)
```

### Removing additional unwanted columns

Remove the column 'number of rounds' as this measures the number of rounds of the particular fight (title bouts are usual 5 round fights, non-title bouts are usually 3 round fights);we are predicting based on past data, not current data.

Remove the column total time fought, as there is already a column total_rounds_fought that measures the same feature

Remove the columns, REV (number of reversal landed) and PASS (number of times a fighter's guard was passed) as they are too specific measures and contain too many values of 0.

```python
#remove the column 'number of rounds', we are predicting based on past data, not current data
raw_data.drop(['no_of_rounds'], axis=1, inplace = True)

#total rounds fought and total time fought both essentially measure the same feature
raw_data.drop(['B_total_time_fought(seconds)', 'R_total_time_fought(seconds)'], axis=1, inplace = True)

#PASS is no. times the guard was passed, REV is the no. of Reversals landed
raw_data.drop(['R_avg_PASS', 'B_avg_PASS', 'R_avg_REV', 'B_avg_REV'], axis=1, inplace = True)
```

### Creating new column for striking accuracy
create a new column for each fighter measuring striking accuracy going into the fight as total number of strikes landed / total number of strikes attempted

```python
#create new column for average total strikes percentage landed
blue_total_att = raw_data['B_avg_TOTAL_STR_att']
blue_total_landed = raw_data['B_avg_TOTAL_STR_landed']
red_total_att = raw_data['R_avg_TOTAL_STR_att']
red_total_landed = raw_data['R_avg_TOTAL_STR_landed']
blue_total_pct = []
red_total_pct = []

for i in range(len(blue_total_att)):
    att = blue_total_att[i]
    landed = blue_total_landed[i]
    if (att == 0):
        blue_total_pct.append(0)
    else:
        blue_total_pct.append(landed / att)

for i in range(len(red_total_att)):
    att = red_total_att[i]
    landed = red_total_landed[i]
    if (att == 0):
        red_total_pct.append(0)
    else:
        red_total_pct.append(landed / att)

raw_data['B_avg_TOTAL_STR_pct'] = blue_total_pct
raw_data['R_avg_TOTAL_STR_pct'] = red_total_pct
```

### Cleaning the Winner column
Update the Winner column to a binary outcome variable
   - Red fighter winning = 1
   - Blue fighter winning = 0

```python
# Cleaning the Winner column (red fighter = 1; blue fighter = 0)
new_winners = []

raw_data = raw_data[raw_data['Winner'] != 'Draw']

raw_data = raw_data.reset_index()

winners = raw_data['Winner']

for i in range(0, len(winners)):
    if ("Red" == winners[i]):
        new_winners.append(1)
    elif ("Blue" == winners[i]):
        new_winners.append(0)

raw_data['Winner'] = new_winners
```

### Creating the column for genders
Add a new column for Gender of fighters
   - M = Male fighter
   - F = Female fighter

```python
#add column for gender
gender = []
weight_class = raw_data['weight_class']

for i in weight_class:
    if 'Women' in i:
        gender.append('F')
    else:
        gender.append('M')

raw_data['Gender'] = gender

raw_data = raw_data.drop(columns=['level_0'])
```

### Creating winning percentage column

For each fight, calculate the winning percentage of each fighter going into the fight. This can be calculated by the total number of wins / total number of fights prior to that particular fight.

```python
#add column for win percentage 
blue_wins = raw_data['B_wins']
blue_losses = raw_data['B_losses']
blue_draws = raw_data['B_draw']
red_wins = raw_data['R_wins']
red_losses = raw_data['R_losses']
red_draws = raw_data['R_draw']
blue_win_pct = []
red_win_pct = []

for i in range(len(blue_wins)):
    wins = blue_wins[i]
    losses = blue_losses[i]
    draws = blue_draws[i]
    total = wins + losses + draws
    if (total == 0):
        blue_win_pct.append(0)
    else:
        blue_win_pct.append(wins / total)

for i in range(len(red_wins)):
    wins = red_wins[i]
    losses = red_losses[i]
    draws = red_draws[i]
    total = wins + losses + draws
    if (total == 0):
        red_win_pct.append(0)
    else:
        red_win_pct.append(wins / total)

raw_data['B_win_pct'] = blue_win_pct
raw_data['R_win_pct'] = red_win_pct
```

### Creating height, weight, reach differential columns

For each fight, calculate the height, weight and reach differential of the fighters. For these categories, the differential will be calculated with respect to the red fighter (favorite). For example, height differential for a fight would be calculated as the red fighters height advantage or disadvantage: height differential = red height - blue height. The same can be applied to weight and reach differential

```python
#add column for height, weight, reach differential 
blue_height = raw_data['B_Height_cms']
blue_weight = raw_data['B_Weight_lbs']
blue_reach = raw_data['B_Reach_cms']
red_height = raw_data['R_Height_cms']
red_weight = raw_data['R_Weight_lbs']
red_reach = raw_data['R_Reach_cms']
height_diff = []
weight_diff = []
reach_diff = []

for i in range(len(blue_height)):
    height_diff.append(red_height[i] - blue_height[i])
    weight_diff.append(red_weight[i] - blue_weight[i])
    reach_diff.append(red_reach[i] - blue_reach[i])

raw_data['height_diff'] = height_diff
raw_data['weight_diff'] = weight_diff
raw_data['reach_diff'] = reach_diff
```

## Machine Learning & Analysis

### Cleaning the dataset for logistic regression
In this next section, we will perform logistic regression algorithms to predict winner of UFC fights
winner is a binary variable:
   - 1: red is the winner of the fight
   - 0: blue is the winner of the fight

First, we need to drop the following non-numeric columns: 'R_fighter', 'B_fighter', 'Referee', 'date', 'location', 'title_bout', 'weight_class','R_Stance', 'B_Stance', 'level_0', 'index', 'gender'

```python
#y variable: 1 (red corner winner of fight), 0: (blue corner winner of fight)

#first drop non-numeric columns
#Winner column (predicted variable) is alredy 0,1 encoded

to_remove = [
    'R_fighter', 'B_fighter', 'Referee', 'date', 'location', 'title_bout', 'weight_class',
    'R_Stance', 'B_Stance', 'index', 'Gender'
]

raw_data.drop(to_remove, axis=1, inplace = True)

print(len(list(raw_data.columns)))
```

### Feature selection
After removing non-numeric columns, our dataset still has 58 features. We will use feature selection to select the 20 most important features to improve the accuracy of our training algorithms.

To help in feature selection, we will implement [Recursive Feaeture Elimination](https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.RFE.html). The basic concept is to recursively construct a model, removing either the worst or best feature at each step. The newton-cg method is used for the solver.

The RFE method has selected the following 20 most important features out of the dataset: 'B_current_win_streak', 'B_avg_SIG_STR_pct', 'B_avg_SUB_ATT', 'B_avg_TD_att', 'B_total_title_bouts', 'B_win_by_KO/TKO', 'B_win_by_Submission', 'B_wins', 'R_avg_SIG_STR_att', 'R_avg_SIG_STR_landed', 'R_avg_TD_landed', 'R_avg_TD_pct', 'R_avg_TOTAL_STR_att', 'R_avg_TOTAL_STR_landed', 'R_losses', 'R_total_title_bouts', 'R_age', 'B_win_by_Decision', 'B_avg_TOTAL_STR_pct', 'R_avg_TOTAL_STR_pct'

```python
#feature selection
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split

#retrieve dependent(y) and independent (X) column(s) from the dataframe
y = raw_data['Winner']
X = raw_data.drop('Winner',axis=1)

#select 20 most important features
logreg = LogisticRegression(solver = 'newton-cg')
rfe = RFE(logreg, 20)
rfe = rfe.fit(X, y)

mask = rfe.support_
columns = list(raw_data.columns)
columns.remove('Winner')
selected_columns = []

#remove unselected features
for i in range(len(mask)):
    if mask[i]:
        selected_columns.append(columns[i])
```

### Statistical analysis
To run statistical analysis on the correlation of each of the features we will use [StatsModels api](https://www.statsmodels.org/stable/index.html).From the above statistical analysis, 2 features (B_avg_SUB_ATT, R_avg_TD_pct, B_win_by_KO/TKO) have p-values > 0.05, 
so we will remove those. 

Now we are left with the following 17 features: 
'B_current_win_streak', 'B_avg_SIG_STR_pct', 'B_avg_TD_att', 'B_total_title_bouts', 'B_win_by_Submission', 'B_wins', 'R_avg_SIG_STR_att', 'R_avg_SIG_STR_landed', 'R_avg_TD_landed', 'R_avg_TOTAL_STR_att', 'R_avg_TOTAL_STR_landed', 'R_losses', 'R_total_title_bouts', 'R_age', 'B_win_by_Decision', 'B_avg_TOTAL_STR_pct', 'R_avg_TOTAL_STR_pct

```python
#statistical analysis
import statsmodels.api as sm

#select only the top 20 features (columns)
X = raw_data[selected_columns]

logit_model=sm.Logit(y,X)
result = logit_model.fit()
print(result.summary())

selected_columns.remove('B_avg_SUB_ATT')
selected_columns.remove('R_avg_TD_pct')
selected_columns.remove('B_win_by_KO/TKO')

print(list(selected_columns))
```

### Model Accuracy
We implemented three different regression algorithms: regularlized logistic regression, K-nearest neighbours classifier and Linear Discriminant Analysis (LDA). LDA had the highest accuracy of about 67%. This is a slight improvement on 64% (the accuracy by simply choosing the favoured fighter), so it represents somewhat of an improvement.

```python
from sklearn.metrics import accuracy_score
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

#select only selected columns
X = raw_data[selected_columns]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

logreg = LogisticRegression()
logreg.fit(X_train, y_train)

y_pred = logreg.predict(X_test)

knn = KNeighborsClassifier(n_neighbors=19)
knn_scores = cross_val_score(knn, X, y, cv=10, scoring='accuracy')

lda = LinearDiscriminantAnalysis()
lda_scores = cross_val_score(lda, X, y, cv=10, scoring='accuracy')


print("Sklearn regularized Logistic Regression accuracy: %0.2f" % (metrics.accuracy_score(y_test, y_pred)))
print("K-nearest neighbours classifier accuracy: %0.2f (+/- %0.2f)" % (knn_scores.mean(), knn_scores.std() * 2))
print("Linear Discriminant Analysis (LDA) classifier accuracy: %0.2f (+/- %0.2f)" % 
      (lda_scores.mean(), lda_scores.std() * 2))
print()
```

Below shows the confusion matrix based on sklearns logistic regression algorithm, this includes correct predictions and incorrect predictions (false negatives and false positives).
The logistic regression algorithm had an accuracy of 0.653276955602537 and precision of 0.6740467404674046.
